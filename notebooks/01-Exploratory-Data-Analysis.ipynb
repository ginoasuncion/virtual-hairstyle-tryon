# Exploratory Data Analysis of Barbershop Model Dataset

This notebook delves into the image dataset utilized by Zhu et al. (2021) in developing the Barbershop model, with a specific focus on hairstyle transfers. It encompasses a collection of images curated by the authors to represent a diverse array of appearances and hair shapes. Among these images, they specifically selected 198 pairs for conducting hairstyle transfer experiments. Through this notebook, we can gain explore and analyze the characteristics of the images that served as training data for the model.

The following properties were calculated for each image.

1. **Total Pixels:** Represents the total number of pixels in each image, providing insight into its size or resolution.
2. **Average Intensity:** Indicates the mean pixel brightness across all channels, reflecting the overall brightness of the image.
3. **Average Sharpness:** Measures the average sharpness of the image, which correlates with clarity and focus.
4. **Average Noise:** Reflects the average level of noise present in the image, capturing random variations or distortions.
5. **Average Vibrancy:** Represents the average color saturation of the image, conveying the richness and intensity of its colors.
